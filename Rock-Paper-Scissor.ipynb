```python
# قم بإنشاء أرقام عشوائية (صور) تستخدمها لاختيار تحركات الذكاء الاصطناعي في اللعبة.
import random
# يتم استخدامه للتعامل مع منطق المؤقت، مما يساعدك على إدارة العد التنازلي وتدفق اللعبة.
import time
# the primary library for computer vision 
# using it for capturing video input, processing images, displaying results.
#  تشغيل الكاميرا، التعرف على الوجوه، أو تعديل الصور. او عرض النتيجه
import cv2

# makes it easier to handle certain tasks يسهل التعامل مع مهام معينه
# like overlaying images (which you are using for displaying AI’s hand move).تراكب الصور
#  تستخدمه لعرض حركة يد الذكاء الاصطناعي
# run image processing and AI functions  تشغيل وظائف معالجة الصور والذكاء الاصطناعي
import cvzone

# يستخدم هذا التطبيق للكشف عن حركات اليد في الوقت الفعلي عبر كاميرا الويب.
# ويساعد في التعرف على حركة يد اللاعب (حجر، ورقة، مقص).
# تتبع ليد اللاهب
from cvzone.HandTrackingModule import HandDetector
```

```python
# 1-  يقوم بتشغيل الكاميرا 

# OpenCV يُهيئ كاميرا الويب باستخدام 
# يشير الرقم 0 إلى أول كاميرا متصلة (عادةً ما تكون كاميرا الويب الافتراضية على جهاز الكمبيوتر).
### ---> if you have multiple cameras connected to your computer , you can change the value
cap = cv2.VideoCapture(0)


# 2 - تحديد أبعاد الفيديو العرض و الارتفاع


#-------> set(property ID  , property value)

# property ID width : 3  ---> openCV الكود الخاص بالعرض في 
# property value : حدد عرض الفيديو إلى 640 بكسل. video frame width

cap.set(3, 640)

# property ID height : 4   openCV الكود الخاص بالارتفاع في 
# property value : حدد ارتفاع الفيديو إلى 640 بكسل. video frame height

cap.set(4, 480)
###############  the frame size for the video capture will be 640x480.  ###############


# 3 - Creates a hand detection object using cvzone. اعداد كائن لكشف اليد

# HandDetector : used to detect hands in the frame.
# maxHands = 1 : means only one hand will be tracked at a time يعني أنه سيتعرف على يد واحدة فقط.
detector = HandDetector(maxHands=1)


# 4 - عمل متغير لتخزين الوقت 
# غالبًا يُستخدم للعد التنازلي قبل اختيار اليد

timer = 0

# 5 - حاله اللعيه 

#لان في بدايه اللعبه لا يكون هناك حاله للعبه false حاله اللعبه في البدايه تكون 
# true : run
# false : stop
# في حاله ان الكمبيونر لسه معرضش نتيجه و العد التانازلي شغال
stateResult = False

# يحدد المتغير ده ما اذا كان اللعبه بدأت ام لا 
# ضغطنا علي زر التشغيل و لا لا 
# true : عندما يبدأ العد التنازلي أو اكتشاف اليد.
startGame = False

# 6 - عمل متغير لتخزين نتيجه اللاعب و الذكاء الاصطناهي

# تحتوي القائمة على نتائج الذكاء الاصطناعي واللاعب. القيمة الأولى للذكاء الاصطناعي، والثانية للاعب.
scores = [0, 0]  # [AI, Player]
```

```python
######## لو بدأنا اللعبه  طول ما اللعبه شغاله#########

while True:
  
  # 1. خلفية اللعبة
  
  #  loads the background image for your game تحميل خلفية ثابتة للعبة (صورة واجهة اللعبة).
  
    imgBG = cv2.imread("BG.png")
    
    # 2. التقاط صورة من الكاميرا
    
    # The variable img contains the image captured from the webcam.
    # success : يعبر عن هل نجحت الكاميرا في التقاط اطار ام لا
    # img variable لو نجح هيتم تخزين الاطار في 
    # If not, success will be False.
    
    success, img = cap.read()
     
    # 3. تصغير الصورة وتحديد منطقة معينة
     
     # بيعمل تغيير لحجم الصوره الي مأخوذه من ال الاطار الي 87% من حجمها الاصلي
     # ثم قصها لتركّز على منطقة معينة ( الجزء الذي تظهر فيه اليد فقط).
     
    imgScaled = cv2.resize(img, (0, 0), None, 0.875, 0.875) # height # width
    
    # المعالجه
    
    # بالنسبه للصور هو بيأخد الصوره بكل الصفوف بتاعتها من البكسلات 
    # لكن من الاعمده بيقص 80 عمود من الشمال 
    # يعني بيشيل الجزاء غير الضروريه من الاطار
    
    imgScaled = imgScaled[:, 80:480] 
    
    # 4. اكتشاف اليد
    # cvzone : تستخدم لاكتشاف اليد وتحديد موقعها وأصابعها.
    
    
    # Find Hands in imgScaled images
    # img : نفس الصورة مع الأيدي المرسومة عليها، للحصول على ردود فعل بصرية
    hands, img = detector.findHands(imgScaled)  # with draw
    
    # 1- لو بدأنا اللعبه 
    
    if startGame :
      
      if stateResult is False :  
        timer = time.time() - initialTime 
        cv2.putText(imgBG , str(int(timer)) , (605 , 435) , cv2.FONT_HERSHEY_PLAIN , 6 , (255 , 0 , 255) , 4 )
        
        # بعد 3 ثواني
        if timer > 3 :
          stateResult = True # AI --> result
          timer = 0
          
          # في حاله اكتشاف يد
          if hands :
            
            playerMove = None 
            
            hand = hands[0]
            
            # PLAYER 
            fingers = detector.fingersUp(hand)
            if fingers == [0 , 0 , 0 , 0 , 0]:
              playerMove = 1 # rock
            if fingers == [1 , 1 , 1 , 1 , 1]:
                playerMove = 2 # paper
            if fingers == [0 , 1 , 1 , 0 , 0]:
              playerMove = 3 # scissors
              
              # AI
            randomNumber = random.randint(1,3)
            imgAI = cv2.imread(f'{randomNumber}.png' , cv2.IMREAD_UNCHANGED)
            imgBG = cvzone.overlayPNG(imgBG , imgAI , (149 , 310))
            
            # player win 
            if (playerMove == 1 and randomNumber == 3) or \
              (playerMove == 2 and randomNumber == 1) or \
                (playerMove == 3 and randomNumber == 2) :
                
              scores[1] += 1
              
            # AI win 
            if (playerMove == 3 and randomNumber == 1) or \
              (playerMove == 1 and randomNumber == 2) or \
                (playerMove == 2 and randomNumber == 3 ) :
                  
              scores[0] += 1
              
    imgBG[234:654 , 795:1195] = imgScaled
    
    if stateResult :
      imgBG = cvzone.overlayPNG(imgBG , imgAI , (149 , 310))
      
    cv2.putText(imgBG , str(scores[0]) , (410 , 215) , cv2.FONT_HERSHEY_PLAIN , 4 , (255 , 255 , 255) , 6)
    cv2.putText(imgBG , str(scores[1]) , (1112 , 215) , cv2.FONT_HERSHEY_PLAIN , 4 , (255 , 255 , 255) , 6)
    
    cv2.imshow("Game" , imgBG)
    
    key = cv2.waitKey(1)
    
    if key == ord('s'):
      startGame = True
      initialTime = time.time()
      stateResult = False
  
              
              
            
            
            
            
            
            
            
            
            
            
    
    
```
